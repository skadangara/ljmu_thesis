```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("data/input_data.xlsx")
```

```python
# Univariate Analysis
```

```python
#Distribution of Distance and Emission

sns.histplot(df['distance'], bins=50, kde=True)
plt.title("Distribution of Distance (km)")
plt.show()

sns.histplot(df['emission'], bins=50, kde=True)
plt.title("Distribution of Emission (kg)")
plt.show()
```

```python
# Count Plots for Categorical Features

sns.countplot(x='shipping_mode', data=df)
plt.xticks(rotation=45)
plt.title("Shipping Mode Counts")
plt.show()

sns.countplot(x='transportation_mode', data=df)
plt.xticks(rotation=45)
plt.title("Transportation Mode Counts")
plt.show()
```

```python
# Bivariate Analysis
```

```python
# Emission vs Distance

sns.scatterplot(x='distance', y='emission', hue='transportation_mode', data=df)
plt.title("Emission vs Distance by Transport Mode")
plt.show()
```

```python
# Boxplot of Emission per Transport Mode
sns.boxplot(x='transportation_mode', y='emission', data=df)
plt.xticks(rotation=45)
plt.title("Emission by Transportation Mode")
plt.show()
```

```python
# Grouped Bar / Aggregated Plots
```

```python
# Average Emission by Mode

df.groupby('transportation_mode')['emission'].mean().sort_values().plot(kind='barh')
plt.title("Average Emission by Transportation Mode")
plt.show()
```

```python
# Group by shipping_mode and compute average emission
avg_emission = df.groupby('shipping_mode')['emission'].mean().sort_values()

# Bar plot
avg_emission.plot(kind='bar', color='skyblue')
plt.ylabel("Average Emission")
plt.title("Average Emission by Shipping Mode")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
sns.boxplot(x='shipping_mode', y='emission', data=df)
plt.ylabel("Emission")
plt.title("Emission Distribution by Shipping Mode")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Replace this with your actual data loading
# df = pd.read_csv("your_data.csv")

# Dummy data (for illustration only  remove if using your real dataset)
import numpy as np
np.random.seed(42)
df = pd.DataFrame({
    'distance_km': np.random.gamma(2, 2000, size=1000),
    'emission': np.random.gamma(2, 150, size=1000),
    'transportation_mode': np.random.choice(['Air', 'Truck', 'Rail', 'Ship'], size=1000),
    'shipping_mode': np.random.choice(['Standard', 'Express', 'Overnight', 'Economy'], size=1000)
})

# Set Seaborn theme (light mode, clean for thesis)
sns.set_theme(style="whitegrid")

# --------- PLOT 1: Distribution of Emission -----------
plt.figure(figsize=(8, 5))
sns.histplot(df['emission'], bins=50, kde=True, color='orange')
plt.title("Distribution of Emission (g CO)", fontsize=14)
plt.xlabel("Emission (grams)", fontsize=12)
plt.ylabel("Shipment Count", fontsize=12)
plt.tight_layout()
plt.savefig("plot_emission_distribution.png", dpi=300)
plt.show()

# --------- PLOT 2: Distance vs Emission by Transportation Mode -----------
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='distance_km', y='emission', hue='transportation_mode', alpha=0.6)
plt.title("Distance vs Emission by Transportation Mode", fontsize=14)
plt.xlabel("Distance (km)", fontsize=12)
plt.ylabel("Emission (g CO)", fontsize=12)
plt.legend(title="Mode", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig("plot_distance_vs_emission.png", dpi=300)
plt.show()

# --------- PLOT 3: Emission by Transportation Mode (Boxplot) -----------
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='transportation_mode', y='emission', palette='Set2')
plt.title("Emission by Transportation Mode", fontsize=14)
plt.xlabel("Transportation Mode", fontsize=12)
plt.ylabel("Emission (g CO)", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("plot_emission_by_transport_mode.png", dpi=300)
plt.show()

# --------- PLOT 4: Average Emission by Shipping Mode -----------
plt.figure(figsize=(8, 5))
avg_emission = df.groupby('shipping_mode')['emission'].mean().sort_values()
avg_emission.plot(kind='bar', color='slateblue')
plt.title("Average Emission by Shipping Mode", fontsize=14)
plt.ylabel("Average Emission (g CO)", fontsize=12)
plt.xlabel("Shipping Mode", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("plot_avg_emission_by_shipping_mode.png", dpi=300)
plt.show()
```
